```python
#1) Explore more regex patterns Eg. The regex pattern used to validate email addresses, mobile no, string, and more
# for email
import re
email = input("Enter Email: ")
pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
# for mobile number
mobile = input("Enter your mobile number: ")
pattern1 = r'^[6-9]\d{9}$'
if re.match(pattern1,mobile):
    print("Valid mobile number")
else:
    print("Invalid mobile number")
```

```python
# for string only alphabet
name = input("Enter your name: ")
pattern2 = r'^[A-Za-z]+$'
if re.match(pattern2,name):
    print("Valid name string")
else:
    print("Invalid name string")
```

```python
# for password validation 
password = input("Enter your password: ")
pattern3 = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&]).{8,}&'
if re.match(pattern3,password):
    print("Password is valid and strong")
else:
    print("Password is valid but weak")
```

```python
#2) Explore more datetime function and uses in pandas 
import pandas as pd
date = pd.to_datetime("01-12-2006")   # no matter how you descibe your date it will take it as yyyy-mm-dd
print(date,type(date))

#converting a column to date time
df = pd.DataFrame({
    'Date':['01-01-2026','02-05-2026','11-02-2026']
})
df['Date'] = pd.to_datetime(df['Date'])
print(df)

current = pd.Timestamp.now()
print(current)

#Extrach date component
df1 = pd.DataFrame({
    'Date':pd.to_datetime(['2026-01-12','2026-02-14','2026-05-18'])
})
df1['Year'] = df1['Date'].dt.year
df1['Month'] = df1['Date'].dt.month
df1['Day'] = df1['Date'].dt.day
print(df1)

df1['Day_name'] = df1['Date'].dt.day_name()
df1['Month_name'] = df1['Date'].dt.month_name()
print(df1)

#sum of monthly sales 
df2 = pd.DataFrame({
    'Sales':[122,342,132,534]
},index=pd.to_datetime(['2026-01-01','2026-01-23','2026-02-02','2026-02-08']))
monthly = df2.resample('ME').sum()
print(monthly)
```

```python
#3) Import an meaningful csv file for data analysis and perform data cleaning, and analysis for meaningful output
# performing meaningfull analysis on real state csv file.
import pandas as pd 
import numpy as np

# 1. loading csv file 
data = pd.read_csv("real_estate_data.csv")
print(data.head())

print("\n",data.shape)   # gives the rows x column 
print("\n",data.columns) # returns the list of all the column fields
print("\n",data.dtypes)  # returns the datatype of all the columns feilds
print("\n",data.info)    # this prints all the info od the csv file

# checking for missing values
print(data.isnull().sum()) # this returns the number of missing values in each column
# filling null values with median values 
data.fillna(data.median(numeric_only=True),inplace=True)
print("\n",data.isnull().sum()) 

# checking for duplicates 
print(data.duplicated().sum()) # our data contain no duplicate values

# basic statistics summary
print(data.describe(),"\n")

avg_price = data['Price_Lakhs'].mean()
print("Average price:",avg_price,"\n")

print(data.loc[data['Price_Lakhs'].idxmax()])  # shows the most expensive property
print(data.loc[data['Price_Lakhs'].idxmin()])  # showes the cheapest property

print(data.corr(numeric_only=True))  # this returns the correlation matrix of the numeric column 
```
