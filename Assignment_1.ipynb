```python
# Question 1
#Write a python program that takes in a student name,class.It should also take in five subject marks of the students and find the total mark 
#and percentage.Display the result in such a way that their name,class,and percentage are print.

name = input("Enter you name: ")
classs = int(input("Enter the class in which you study: "))
sub1 = int(input("Enter the marks of 1st subject: "))
sub2 = int(input("Enter the marks of 2nd subject: "))
sub3 = int(input("Enter the marks of 3rd subject: "))
sub4 = int(input("Enter the marks of 4th subject: "))
sub5 = int(input("Enter the marks of 5th subject: ")) 
percentage = (sub1+sub2+sub3+sub4+sub5)/500*100
print(percentage)
print(f"The marks of the sutdent name {name} of class {classs} have percentage {percentage}")
```

```python
# Question 2
#input 2 strings and concatinate them and store in another vairable. then perform the methods on it 

str1 = input("Enter your 1st string: ")
str2 = input("Enter your 2nd string: ")
finalstr = str1+" "+str2
print(str1)
print(str2)
print(finalstr)
print(str.lower(finalstr))
print(str.upper(finalstr))
print(str.title(finalstr))
print(str.swapcase(finalstr))
print(str.capitalize(finalstr))
print(str.casefold(finalstr))
print(finalstr.center(80))
print(finalstr.count("p"))
print(finalstr.endswith("e"))
print(finalstr.find("i"))
print(finalstr.isalnum())
print(finalstr.isdigit())
print(finalstr.isnumeric())
print(finalstr.isspace())
print(finalstr.replace("i","u"))
```
