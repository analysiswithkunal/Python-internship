```python
#1) Practise Pandas Series Create a Pandas Series from Dictionary Create a Pandas Series from Lists Access the elements of a Series in Pandas
# creating a pandas series from a dictionary
import pandas as pd
import numpy as np
visitor = {
    'Day1':120,
    'Day2':135,
    'Day3':150,
    'Day4':145
}
s1 = pd.Series(visitor)
print(s1,"\n")
# creating pandas series form list 
l1 = ["Jaipur","Varanasi","Hyedrabaa"]
s2 = pd.Series(l1,index=[11,12,13])
print(s2)
# accessing element form series
print(s1['Day1'])
print(s1['Day2']+s1['Day4'])
print(s2[11])
```

```python
#2) DataFrames Make a Pandas DataFrame with a two-dimensional Python list Create DataFrame from Python dict Create Pandas 
# dataframe using list of lists Create a Pandas dataframe using list of tuples Create a Pandas DataFrame from List of Dicts.
# creating pandas data frame from two series
data = {
    "Name":["Kunal","Khushi","Krishna","Harsh","Dhruv"],
    "Age":[20,21,19,18,100]
}
df = pd.DataFrame(data)
print(df,"\n")
print(df.loc[0])    # loc[] function is used to return rows in a dataframe
#creating dataframe with list of lists
sample_data = [
    ["Name","Age"],
    ["Kunal",20],
    ["Khushi",21]
]
df1 = pd.DataFrame(sample_data,index=[0,12,18])
print(df1)
# creating dataframe with lists of dictionary
raw_data = [
    {'City':'Jaipur','people':'20M'},
    {'City':'Udaipur','people':'18M'}
]
df2 = pd.DataFrame(raw_data)
print(df2)
```

```python
#3)Data iteration Different ways to iterate over rows in Pandas Dataframe Selecting rows in pandas DataFrame based on conditions Select any row from 
# a Dataframe using iloc[] Limited rows selection with given column Drop rows from the dataframe based on certain condition applied on a column Insert 
# row at given position in Pandas Dataframe Create a list from rows in Pandas dataframe.

# using loc[] 
print(df.loc[0,'Age'])    # loc[] takes index and column labels
print(df.loc[3])
print(df.loc[:,'Name'])    # it returns the name column along with index

#using iloc[] 
print(df.iloc[0,1])        # iloc[] takes integer arguments and treat them as position
print(df2.iloc[1,1])

# selection of rows based on condition
print(df[df['Age']>19])
df.query('Age>19')      # by using query function

# inserting row at perticular position in dataframe
new_row = pd.DataFrame({'Name':["Manas"],'Age':[37.5]})
pos = 4
df = pd.concat([df.iloc[:pos],new_row,df.iloc[pos:]],ignore_index=True)
print(df)

#creating list from row of pandas dataframe
print(type(df))
print(df[2])
```
