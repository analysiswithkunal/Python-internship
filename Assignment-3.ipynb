```python
# Question 2
#write a function for basic airthmatic operation by taking two inputs from users
def add():
    a=int(input("Enter the first number: "))
    b=int(input("Enter the second number: "))
    r=a+b
    print("the result is :",r)

def sub():
    a=int(input("Enter the first number: "))
    b=int(input("Enter the second number: "))
    r=a-b
    print("the result is :",r)

def mul():
    a=int(input("Enter the first number: "))
    b=int(input("Enter the second number: "))
    r=a*b
    print("the result is :",r)

def divide():
    a=int(input("Enter the first number: "))
    b=int(input("Enter the second number: "))
    r=b/a
    print("the result is :",r)

def modulo():
    a=int(input("Enter the first number: "))
    b=int(input("Enter the second number: "))
    r=a%b
    print("the result is :",r)

def floor_division():
    a=int(input("Enter the first number: "))
    b=int(input("Enter the second number: "))
    r=a//b
    print("the result is :",r)
    
value=int(input(f"\n 1.add \n 2.sub \n 3.multiply \n 4.divide \n 5.modulo \n 6.floor division"))
match value:
    case 1:
        add()
    case 2:
        sub()
    case 3:
        mul()
    case 4:
        divide()
    case 5:
        modulo()
    case 6:
        floor_division()
    case _:
        print("Invalid choice")
```

```python
# Question 3
#write a function in python to check a number if it is a pallindrome or not.
def check_palli():
    num=input("enter you number: ")
    rev_num = num[-1::-1]
    if(rev_num == num):
        print("Entered number is pallindrome.")
    else:
        print("Number is not a pallindrome.")
        
check_palli() 
```
