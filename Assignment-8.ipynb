```python
#1) Convert a series of date-strings to a timeseries?
import pandas as pd

date_series = pd.Series([
    '01 jan 2019',
    '10-02-2016',
    '20180207',
    '2014/12/12',
    '2015-05-03',
    '2021-04-06T11:11'
])
print('Original Series:')
print(date_series)

print("\n converted series:")
timeseries = pd.to_datetime(date_series)
print(timeseries)
```

```python
#2)Create two DataFrames, df1 and df2, with a common column (e.g., 'ID'). 
#>Perform an inner merge on this common column and display the resulting DataFrame
df1 = pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['Kunal','Khushi','Krishna','Harsh']
})
df2 = pd.DataFrame({
    'ID':[3,4,5,6],
    'Age':[23,34,23,12]
})
print('DataFrame 1:\n',df1)
print('DataFrame 2:\n',df2)
merge_df = pd.merge(df1,df2,on='ID',how='inner')
print('merged dataframe :\n',merge_df)

#>Perform a left join of df1 and df2 on the 'ID' column. 
# Explain how missing values are handled in the resulting DataFrame. 
# Right Join and Index-Based Join. 
leftmerge = pd.merge(df1,df2,on='ID',how='left')
print('Left merged dataframe :\n',leftmerge)  # only values present in the df2 will inserted on df1 and df1 is majour dataframe
rightmerge = pd.merge(df1,df2,on='ID',how='right')
print('Right merged dataframe :\n',rightmerge) # in this only values present in the df1 will inserted in the df2 and df2 will be the majour dataframe

# indexed based joins
df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')
index_left = df1_indexed.join(df2_indexed,how='left')
print(index_left)
index_right = df1_indexed.join(df2_indexed,how='right')
print(index_right)
```

```python
# joing dataframe with multiple keys/column
df3 = pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['Kunal','Khushi','Krishna','Harsh'],
    'Year':[2020,2023,2022,2021]
})
df4 = pd.DataFrame({
    'ID':[3,4,5,6],
    'Age':[23,34,23,12],
    'Year':[2022,2021,2022,2021]
})
multikeymerge = pd.merge(df3,df4,on=['Year','ID'],how='inner')
print('Multi-key merge: \n',multikeymerge)
```
