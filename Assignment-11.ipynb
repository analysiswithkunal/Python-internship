```python
#1) Combining a one and a two-dimensional NumPy Array.
import numpy as np 
arr1 = np.array([[1,2,3],[4,5,6],[11,12,13]])
arr2 = np.array([7,8,9])
print(arr1,"\n")
print(arr2)
result1 = np.vstack((arr1,arr2))
print("\n",result1)   # combines the arrays vertically and or row wise

result2 = np.column_stack((arr1,arr2))
print("\n",result2)  # this inserts the array as new column in the given array

result3 = np.concatenate((arr1,arr2.reshape(1,3)), axis=0)   # this also combines the array row wise but it first reshape the 1d array
print("\n",result3)

#2) Flatten a 2d numpy array into 1d array
result4 = arr1.reshape(-1)
print("\n",result4)

#3) Reverse a numpy array
result4 = result4[::-1]
print("\n",result4)        # reversing an array after flatting it then again reshape it as required 

result5 = arr1[::-1]
print("\n",result5)       # reversing an array as it is 
```

```python
#4) Practice operations like Get the maximum value from given array Get the minimum value from given array Find the number of rows and columns of a given 
# array using NumPy Select the elements from a given array (each element and specific element) Find the sum of values in a 2D array using for loop Adding, 
# Subtracting, multiplying, dividing arrays in Numpy

arr3 = np.array([[12,324,41],
                 [34,453,23],
                 [89,676,87]])
max_element = np.max(arr3)
print(max_element)
min_element = np.min(arr3)
print(min_element)
print('rows:',arr3.shape[0])  
print('columns:',arr3.shape[1]) 

# accessing elements in the array
for i in arr3:
    for j in i:
        print(j)
        
# accessing the specfic element in the array
print(arr3[2][1])  # arr[row_index][column_index]

# adding array element by for loop
s = 0
for i in arr3:
    for j in i:
        s = s+j
        
print("Final sum of array is:",s)
```

```python
# airthmatic operations on array
t = np.array([
    [[1,2,3,4],[2,3,4,6],[5,6,7,9],[7,4,9,4]],
    [[3,1,6,5],[6,3,4,7],[6,8,4,8],[6,2,3,1]]
])
s = np.array([
    [[1,2,3,4],[2,3,4,6],[5,6,7,9],[7,4,9,4]],
    [[2,1,5,3],[5,6,6,8],[4,8,1,9],[4,2,3,1]]
])
print(t+s)
print(t-s)
print(t/s)
print(t*s)
print(t)
print(s)
print(t**s)
res= np.mean(t,axis=2)   # similarly we can perform sum,median,mode,std,variance
print(res)
```

```python
# 5) Iterate 3D array using for loop and nditer
arr4 = np.array([
    [[1,2,3,4],[2,3,4,6],[5,6,7,9],[7,4,9,4]],
    [[3,1,6,5],[6,3,4,7],[6,8,4,8],[6,2,3,1]]
])
for i in arr4:
    for j in i:
        for k in j:
            print(k) # this returns each element of the array
            
print("\n","traversing using nditer")
            
for t in np.nditer(arr4):   # nditer function traverse all the dimension itself no need of nested loops to increase time complexity
    print(t)
```
