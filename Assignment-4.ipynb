```python
# Question 1
#1. Write a Python function to find the maximum of three numbers.
def max():
    a=int(input("Enter your 1st number: "))
    b=int(input("Enter your 2nd number: "))
    c=int(input("Enter your 3rd number: "))
    if(a>b)and(a>c):
        print(f"{a} is the maximum among {a},{b},{c}.")
    elif(b>a)and(b>c):
        print(f"{b} is the maximum among {a},{b},{c}.")
    else:
        print(f"{c} is the maximum among {a},{b},{c}.")

max()
```

```python
# Question 2
#2. Write a Python function that takes a list and returns a new list with distinct elements from the first list.
def uni_list():
    
    n=int(input("Enter the numbers you want to enter: "))
    lis1=[]
    for i in range(n):
        e=int(input(f"Enter your {i} element: "))
        lis1.append(e)
    print(lis1)
    lis2=list(set(lis1)) # because sets cannot have duplicate values
    print(lis2)

uni_list()
```

```python
# Question 3
#3. Write a Python function to multiply all the numbers in a list.
def mul_list():
    n=int(input("Enter the numbers you want to enter: "))
    lst1=[]
    for i in range(n):
        e=int(input(f"Enter your {i} element: "))
        lst1.append(e)
    print(lst1)
    mul=1
    for i in range(n):
        mul=mul*lst1[i]
    print(f"Multiply of your list is {mul}.")

mul_list()   
```

```python
# Question 4
#4. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.
def factorial(num):
    if(num<0):
        return print("Factorial cannot be claculated.")
    elif(num==0)or(num==1):
        return 1
    else:
        return num*factorial(num-1)

print(factorial(8))
```

```python
# Question 5
# 5. Write a Python program to reverse a string.
def rev_string():
    istring = input("Enter your String: ")
    rstring1 = istring[-1::-1]
    l=str.split(istring)      # It breaks the string and convert each words into the list.
    r_l=l[-1::-1]
    rstring2 = " ".join(r_l)  # It joins the list back to the string format.
    print(f"The reverse string is: {rstring1}.")
    print(f"The another reverse string is: {rstring2}.")
    
rev_string()
```

```python
# Question 6
#6. Write a Python function to check whether a number falls within a given range.
def num_check(start,end,value):
    if(start<=value<=end):
        print("the number lies between the numbers.")
    else:
        print("the number not lies in the given range.")

num_check(2,5,3)
num_check(12,78,50)
num_check(45,56,23)        

    
```

```python
# Question 7
#7. Write a Python function to Print Even Numbers from a Given List
def chk_even():
    n = int(input("Enter the number upto which you want to insert numbers: "))
    ori_list = []
    for i in range(n):
        e=int(input(f"enter the {i}th element: "))
        ori_list.append(e)
    print(ori_list)
    even_list = []
    for i in range(n):
        if(ori_list[i]%2 == 0):
            even_list.append(ori_list[i])
        else:
            pass
    print(f"even element's list is:{even_list}")
    
chk_even()
```

```python
# Question 8
#8. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.
import math
def chk_prime(num):
    if(num<=1):
        return False
    
    if(num==2):
        return True
    
    if(num%2==0):
        return False
    
    limit=int(num/2)
    for i in range(3,limit+1,2):
        if(num%i==0):
            return False
    
    return True                

print(chk_prime(2))
print(chk_prime(0))
print(chk_prime(11))
print(chk_prime(23))
print(chk_prime(7))
print(chk_prime(56))
print(chk_prime(19))
```

```python
# Question 9
#9. Write a Python function that accepts a string and counts the number of upper and lower case letters. 
def count_alphabets():
    ipt_string = input("Enter your string: ")
    countU = 0
    countL = 0
    for i in range(len(ipt_string)):
        if(ipt_string[i].isupper()):
            countU+=1
        elif(ipt_string[i].islower()):
            countL+=1
        else:
            pass
    print(f"The uppercase letters are: {countU}.")
    print(f"The lowercase letters are: {countL}.")
        
count_alphabets()
```
