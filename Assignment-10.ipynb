```python
#1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]] 
import pandas as pd
import numpy as np
arr = np.array([[6,-8,73,-110],[np.nan,-8,0,94]])
print(arr)
#replacing nan with zero
arr[np.isnan(arr)] = 0
print(arr)
arr[:,[0,2]] = arr[:,[2,0]]
arr[:,[1,3]] = arr[:,[3,1]]
print(arr)
```

```python
#2) Move axes of 3D array to new positions
arr2 = np.zeros((3,4,5))
print(arr2)
print(arr2.shape)

result1 = np.moveaxis(arr2,0,-1)
print(result1.shape)

result2 = np.moveaxis(arr2,-1,0)
print(result2.shape)

result3 = np.moveaxis(arr2,[0,1],[1,0])
print(result3.shape)
```
